```python
import json

with open("markets.json", "r") as f:
    markets_data = json.load(f)



for i, line in enumerate(json.dumps(markets_data, indent=2).splitlines()):
    if i >= 100:
        break
    print(line)

```

```python
# All market ids from data
market_ids = [m["id"] for m in markets_data["data"]]
print(f"{len(market_ids)} markets")
for mid in market_ids:
    print(mid)
```

```python
# Store market_ids that don't start with a number
non_numeric_market_ids = [mid for mid in market_ids if not mid[0].isdigit()]
non_numeric_market_ids
```

```python
first_set_market_ids = [mid for mid in market_ids if mid.startswith("1st_set")]
first_set_market_ids
```
